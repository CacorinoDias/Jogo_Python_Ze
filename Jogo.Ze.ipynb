```python
import pygame #importar a bib de jogos
import sys 


pygame.init() #iniciar o pygame

#criar o ecrã agora

comprimento_ecra = 600
largura_ecra = 600


cor_De_Fundo_Preta = (0,0,0)
cor_pontos = (255,255,255)
#os zeros fazem o fundo negro denovo
#variaveis do boneco
cor_boneco = (130,22,0)
#os numeros sao do vermenlho
Posi_Boneco = [300,300]
dim_boneco = [30,30]
vel_boneco = 20

#caraceristicas dos tiros
cor_tiros = ((255,255,0),(51,255,51),(255,0,255),(192,192,192),(0,0,255))
velocidade = 10
tiros_raio = 5


from random import randrange
from random import random
from random import choice

ecra = pygame.display.set_mode((comprimento_ecra,largura_ecra)) 

#no relogio estamos a definir a velocidade com que o jogo acontece
relogio = pygame.time.Clock()

num_tiros = []
#lista que vai armazenar os tiros que vao ser criados

Pontos = 0
#vamos somar a este valor, dentro do loop, 10 sempre que uma bolinha sair do ecrã

myfont = pygame.font.Font('freesansbold.ttf', 35)

#o myfonnt define o tipo de letra que vamos utilizar. Font maiusculo é para ficar em bolt


def nivel(Pontos,velocidade):
    if Pontos < 500:
        velocidade = 5
    elif Pontos < 1500:
        velocidade = 10
    else:
        velocidade = 15
    return velocidade


def coordenadas_tiros(num_tiros):
    quando = random()
    #o que fazemos aqui com este quando é meter um delay, ese so vai fazer o append quando este numero aleatorio for menor que o.1
    if len(num_tiros) < 10 and quando < 0.1:
        Posicao_x = randrange(tiros_raio,largura_ecra-tiros_raio)
        Posicao_y = 0
        num_tiros.append([Posicao_x,Posicao_y])
#função que define de onde os tiros vao sair

def desenho_tiros(num_tiros):
    for tiros_posicao in num_tiros:
        tiros_raio = 5
        pygame.draw.circle(ecra,choice(cor_tiros),(int(tiros_posicao[0]),int(tiros_posicao[1])),tiros_raio)
             #o int está ali porque para os circulos nao podemos ter floats            

def movimento_tiros(num_tiros):
    for tiros_posicao in num_tiros:
        if tiros_posicao[1] >= 0 and  tiros_posicao[1] <= comprimento_ecra:
            tiros_posicao[1] += velocidade
        else:
            tiros_posicao[1] = 0
            tiros_posicao[0] = randrange(tiros_raio,largura_ecra-tiros_raio)
#este if faz cair os blocos de cima o else diz para que quando eles chegarem ao fim sairem denovo de outro lugar qualquer
           

def tiro_certeiro(Posi_Boneco,num_tiros):
    for tiros_posicao in num_tiros:
        
        B_x=Posi_Boneco[0]
        B_y=Posi_Boneco[1]
        t_x=tiros_posicao[0]
        t_y=tiros_posicao[1]
    
    
        if (B_x < (t_x - tiros_raio)) and ((t_x - tiros_raio) < B_x + 30) or (B_x < (t_x + tiros_raio) and ((t_x + tiros_raio) < B_x + 30 )):
            if (B_y < (t_y - 30)) and ((t_y - tiros_raio) < B_y + 30) or (B_y < (t_y + tiros_raio) and ((t_y + tiros_raio) < B_y + 30 )):
                return True
            return False
# o que o True e False faze é: se true a função continua a correr, se false ela para

#vamos definir agora o game loop


final_do_jogo = False

while not final_do_jogo:
    for event in pygame.event.get(): 
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            #este codigo é para fechar o jogo na cruzinha

        if event.type == pygame.KEYDOWN:
            #aqui estamos a determinar que o sistema se entere que a tecla foi primida
            x = Posi_Boneco[0]
            y = Posi_Boneco[1]
            if event.key == pygame.K_a:
                x -= vel_boneco     
            elif event.key == pygame.K_d:
                x += vel_boneco  
            elif event.key == pygame.K_w:
                y -= vel_boneco
            elif event.key == pygame.K_s:
                y += vel_boneco
            Posi_Boneco = [x,y]
       
    ecra.fill(cor_De_Fundo_Preta)
    
    #a linha de cima o que faz é apagar a cor do oneco quando o moveos, para que ela nao cresça        
       
    pygame.draw.rect(ecra,cor_boneco,(Posi_Boneco[0],Posi_Boneco[1],dim_boneco[0],dim_boneco[0]))
#a linha de cima imprime o nosso quadrado
    movimento_tiros(num_tiros)
    
    coordenadas_tiros(num_tiros)
    
    desenho_tiros(num_tiros)
    
    for i in num_tiros:
        if i[1] == 600:
            Pontos += 10
    #aqui definimos um loop para contar 10 pontos cada vez que uma bolinha passar o final
    #as linhas debaixo sao para que este valor apareça na tela.
    text = str(Pontos)
    rotulo = myfont.render(text, 1, cor_pontos)
    ecra.blit(rotulo,(comprimento_ecra-340, largura_ecra-100))
    
    velocidade = nivel(Pontos,velocidade)
        
    if tiro_certeiro(Posi_Boneco,num_tiros):
        Texto_Final = "Game Over" 
        rotulo = myfont.render(Texto_Final, 1, cor_pontos)
        ecra.blit(rotulo,(comprimento_ecra-400, largura_ecra-300))   
        final_do_jogo = True
        
    relogio.tick(30)
    
    pygame.display.update()
    #sempre que fazemos algo no jogo temos que ter o update
```
